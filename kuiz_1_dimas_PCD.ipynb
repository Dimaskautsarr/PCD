## Soal 4 (30 points)
Soal: Kombinasikan Operasi Geometri, Aritmatika, dan Logika pada Gambar
Gunakan dua gambar sebelumnya (gambar_2.png dan gambar_3.png). Tulislah program Python menggunakan OpenCV dan NumPy untuk melakukan langkah-langkah berikut:

1. Operasi Geometri:


*   Lakukan rotasi pada gambar_2.png sebesar 30 derajat searah jarum jam dengan pusat rotasi berada di tengah gambar.
*   Lakukan translasi (pergeseran) pada gambar_3.png dengan menggeser gambar sejauh 10 piksel ke kanan dan 50 piksel ke bawah.


2. Operasi Aritmatika:
*   Gabungkan kedua gambar hasil operasi geometri menggunakan penjumlahan (cv2.add).

3. Operasi Logika:
*   Lakukan operasi NOT (cv2.bitwise_not) pada hasil penjumlahan untuk menghasilkan gambar akhir.



**Instruksi:**

Pastikan kedua gambar memiliki dimensi yang sama sebelum melakukan operasi.
Tampilkan hasil dari setiap langkah operasi: hasil rotasi, hasil translasi, hasil penjumlahan, dan hasil akhir setelah operasi NOT.

Gunakan fungsi matplotlib.pyplot untuk menampilkan gambar hasil setiap tahap dengan menggunakan plt.imshow().

```python
# kolom jawaban
# Load the image
image = cv2.imread('gambar_2.png')
# Define the angle of rotation (in degrees)
angle = 30
# Get the image dimensions
height, width = image.shape[:2]
# Calculate the center of the image
center = (width // 2, height // 2)
# Define the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
# Perform the rotation
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
# Display the original and rotated images
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image")
plt.show()

```

```python
# Load the image
image = cv2.imread('gambar_3.png')
# Define the translation parameters (x and y shifts)
tx = 10 # Shift 75 pixels to the right
ty = 50 # Shift 150 pixels down
# Create the translation matrix
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
# Perform the translation
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))
# Display the original and translated images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title("Translated Image")
plt.show()
```

```python
image_1=cv2.resize(rotated_image,(256,256))
image_2=cv2.resize(translated_image,(256,256))
# hasil gambar Addition
result_add = cv2.add(image_1, image_2)
plt.imshow(result_add)
plt.show()
# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_add.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

```python
# hasil gambar Addition
result_logical = cv2.bitwise_not(result_add,mask = None)
plt.imshow(result_logical)
plt.show()
# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_logical.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```
